```python
import pickle
import matplotlib.pyplot as plt
import datetime

# Define the file path of the pickle file
filename = '/nbhome/Haemah.Akhtar/a68a_pixel_areas2020.pkl'

# Load data from the pickle file
with open(filename, 'rb') as file:
    data = pickle.load(file)

# Initialize lists to store timestamps and areas_ps data
timestamps = []
areas_ps = []

# Process the data
for d in data:
    for filename, value in d.items():
        # Extract the date from the filename
        date_str = filename.split('/')[-1].split('_')[1]
        
        try:
            # Convert the date string to a datetime object
            date = datetime.datetime.strptime(date_str, '%Y-%m-%d').date()
        except ValueError:
            continue
        
        # Append the date and areas_ps value to the respective lists
        timestamps.append(date)
        areas_ps.append(value)

# Filter out the bergs with areas substantially larger than a clean image from the beginning of the timeseries
clean_image_area = areas_ps[0]
threshold_large = clean_image_area * 2  

timestamps_filtered = []
areas_ps_filtered = []

for i, area in enumerate(areas_ps):
    if area <= threshold_large:
        timestamps_filtered.append(timestamps[i])
        areas_ps_filtered.append(area)

# Filter out the bergs substantially smaller than a clean image from the end of the timeseries
threshold_small = clean_image_area * 0.5  

timestamps_final = []
areas_ps_final = []

for i, area in enumerate(areas_ps_filtered):
    if area >= threshold_small:
        timestamps_final.append(timestamps_filtered[i])
        areas_ps_final.append(area)

# Sort the data by timestamps for a smooth curve
sorted_indices = sorted(range(len(timestamps_final)), key=lambda k: timestamps_final[k])
timestamps_final_sorted = [timestamps_final[i] for i in sorted_indices]
areas_ps_final_sorted = [areas_ps_final[i] for i in sorted_indices]

# Create a line plot to visualize the decrease in areas_ps over time
plt.figure(figsize=(10, 6))
plt.plot(timestamps_final_sorted, areas_ps_final_sorted, marker='o', color='b', label='Filtered Data')
plt.axhline(clean_image_area, color='r', linestyle='dashed', label='Initial Clean Image Area')
plt.axhline(threshold_large, color='g', linestyle='dashed', label='Large Threshold')
plt.axhline(threshold_small, color='m', linestyle='dashed', label='Small Threshold')
plt.xlabel('Date')
plt.ylabel('Areas_PS')
plt.title('Polar Stereographic Areas over Time (Filtered)')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()

# Display the plot
plt.show()
```
